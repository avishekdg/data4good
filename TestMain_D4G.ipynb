```python
# Install the Prediction Guard client
! pip install -U predictionguard

# Python Imports
import os
import json

import pandas as pd
from getpass import getpass

pg_access_token = getpass('Enter your Prediction Guard access token: ')
os.environ['PREDICTIONGUARD_TOKEN'] = pg_access_token
#os.environ['PREDICTIONGUARD_URL'] = "https://intel.predictionguard.com"

import predictionguard as pg

# Transcripts
with open('transcripts.json') as f:
  transcripts = json.load(f)

transcripts['2055']
# Test data
test = pd.read_csv('test.csv')
test.head()
# Initialize the transcript
transcript = "During the visit, I examined Mr. Don Hicks, who is 81 years old and presented with a fungal infection. He had dischromic patches, nodal skin eruptions, and skin rash as symptoms. Upon examination, I confirmed the diagnosis of fungal infection. I advised Mr. Hicks to take precautions such as bathing twice a day, using detol or neem in the bathing water, keeping the infected area dry, and using clean cloths. I did not prescribe any medication for him."
transcript

# Create an LLM "prompt" that we will use to extract certain
# information from the transcript.
prompt = f"""### Instruction:
Extract the patient's condition from the below input.

### Input:
{transcript}

### Response:
""".format(transcript=transcript)

# Use Llama 2 to extract the information.
result = pg.Completion.create(
    model="wizardcode",
    prompt=prompt
)

print(json.dumps(
    result,
    sort_keys=True,
    indent=4,
    separators=(',', ': ')
))

! pip install random-word
from random_word import RandomWords
import random
r = RandomWords()
solution_raw = []
for _, row in test.iterrows():
  words = transcripts[str(row['Transcript'])].split(" ")
  num_words = random.randint(1, 5)
  solution_raw.append([
      row['Id'],
      " ".join(random.choices(words, k=num_words))
  ])

solution = pd.DataFrame(solution_raw, columns=["Id", "Text"])
solution.head()
solution.to_csv("predictionguard_submission_1.csv", index=False)
```
